```python
import networkx as nx
import matplotlib.pyplot as plt
import pandas as pd
from datetime import datetime
import numpy as np
%matplotlib inline
G = nx.path_graph(0)
```

```python
data = pd.read_csv('finalCleaning_1 - kopie.csv')
```

```python
data
```

```python
index = data['Unnamed: 0']
columns = ['departure_airport', 'arrival_airport', 'arrival_lat', 'arrival_lat']
```

```python
df = pd.DataFrame(index = index, columns=columns)
```

```python
df['departure_airport']=data['departure_airport']
df['arrival_airport']=data['arrival_airport']
df['arrival_lat']=data['arrival_lat']
df['arrival_lon']=data['arrival_lon']
```

```python
df
```

```python
avg = {}
```

```python
for i in range (0, len(data)):
    dp= df['departure_airport'][i]
    ai = df['arrival_airport'][i]
    if not(dp in avg):
        avg[dp] = {}
    if not (ai in avg[dp]):
        avg[dp][ai] = [df['arrival_lat'][i],df['arrival_lon'][i]]
    
    if i%1000000 == 0:
        print(i)
        
        
        
```

```python
avg
```

```python
for key in avg:
    for val in avg[key]:
        G.add_node(val, pos=(avg[key][val][1], avg[key][val][0]))
for key in avg:
    for val in avg[key]:
        wheight = avg[key][val]
        G.add_weighted_edges_from([(key, val,avg[key][val] )])
```

```python
fig, ax = plt.subplots(1, 1, figsize=(6, 6))
nx.draw_networkx(G, ax=ax, node_size=5,
                 font_size=6, alpha=.5,
                 width=.5)
ax.set_axis_off()
```
