```python
import networkx as nx
import matplotlib.pyplot as plt
import pandas as pd
from datetime import datetime
import numpy as np
%matplotlib inline
G = nx.path_graph(0)
```

```python
data = pd.read_csv('finalCleaning_1 - kopie.csv')
```

```python
data
```

```python
index = data['Unnamed: 0']
columns = ['departure_airport', 'arrival_airport']
```

```python
df = pd.DataFrame(index = index, columns=columns)
```

```python
df['departure_airport']=data['departure_airport']
df['arrival_airport']=data['arrival_airport']
df['departure_lat']=data['departure_lat']
df['departure_lon']=data['departure_lon']
```

```python
df
```

```python
avg = {}
```

```python
df['departure_lat']
```

```python
for i in range (0, len(data)):
    dp= df['departure_airport'][i]
   
    if not(dp in avg):
        avg[dp] = [df['departure_lat'][i],df['departure_lon'][i]]
    
    if i%1000000 == 0:
        print(i)
        
        
        
```

```python
avg
```

```python
edges = {}




edges
```

```python

for i in range (0, len(data)):
    dp= df['departure_airport'][i]
    ai = df['arrival_airport'][i]
    if not(dp in edges):
        edges[dp] = []
    if not (ai in edges[dp]):
        edges[dp] = edges[dp] + [ai]
    if i%10000000 == 0:
        print(i)
        
        
        
        
```

```python
for key in edges:
    G.add_node(key)
for key in edges:
    for val in edges[key]:
        G.add_weighted_edges_from([(key, val, 1 )])
```

```python
nx.draw(G, with_labels=True)
```

```python
fig, ax = plt.subplots(1, 1, figsize=(6, 6))
nx.draw_networkx(G, ax=ax, node_size=5,
                 font_size=6, alpha=.5,
                 width=.5)
ax.set_axis_off()
```

```python
sg = next(nx.connected_component_subgraphs(G))
sg
```

```python
fig, ax = plt.subplots(1, 1, figsize=(6, 6))
nx.draw_networkx(G, ax=ax, with_labels=False,
                 node_size=5, width=.5)
ax.set_axis_off()
```

```python
deg = nx.degree(G)
sizes = [5 * deg[iata] for iata in G.nodes]
```

```python
labels = {iata: iata if deg[iata] >= 20 else ''
          for iata in G.nodes}
```

```python
pos = avg
```

```python
flipped_pos = {node: (x,-y) for (node, (x,y)) in pos.items()}
```

```python


nx.draw_networkx(G, 
                 font_size=12,
                 alpha=0.5,
                 width=.075,
                 node_size=sizes,
                 labels=labels,
                 pos=pos,
                 cmap=plt.cm.autumn)
```
